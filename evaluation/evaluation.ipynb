# Model Evaluation

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data

```python
data = pd.read_csv("./f1-scores.csv")
data.head()
```

## Visualizations

```python
# Sort by score descending
data = data.sort_values(by='score', ascending=False)

# Set Seaborn style
sns.set(style="whitegrid", context="talk")  # 'talk' for publication-ready

# Add color
palette = sns.color_palette("Set2", len(data))

# Plot
plt.figure(figsize=(6, 6))
ax = sns.barplot(x="model", y="score", data=data, palette=palette, hue="model", legend=False)

# Labels and title
ax.set_xlabel("Model", fontsize=14)
ax.set_ylabel("F1 Score", fontsize=14)
ax.set_title("Model Performance Comparison", fontsize=16)

# Rotate x-axis labels
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Save as high-res image
plt.savefig("model_f1_scores.png", dpi=300)
plt.show()
```
